```python
import pandas as pd
df = pd.read_csv('data.csv', encoding='ISO-8859-1')
```

# Data Cleaning: Clean Age Data

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df_valid_ages = df[df['Age'].notna()]
df_valid_ages
```

# Selecting: Age and Rating

```python
age_and_ratings = df_valid_ages[['Age', 'Overall']]
age_and_ratings
```

# Scatter-Graph: Correlation of Age and Rating

```python
import seaborn as sns
sns.set_theme()

graph = sns.scatterplot(x='Age', y='Overall', data=age_and_ratings)
graph
```
